```python
import datetime
from datetime import datetime

def difference_minutos(data1, data2):
    date1 = str(data1)
    date2 = str(data2)
    if(date1 == "NA" or date1 == "nan" or date1 == "NaN") :
        return float("nan")
    if(date2 == 'NA' or date2 == 'nan' or date2 == "NaN") :
        return float("nan")
    d1 = datetime.strptime(date1, "%Y-%m-%dT%H:%M:%SZ")
    d2 = datetime.strptime(date2, "%Y-%m-%dT%H:%M:%SZ")
    return (d2 - d1).total_seconds() / 60
vec = []
for i in range(720532):
    r = difference_minutos(dados_filtrados3.iloc[i]['Partida.Real'], dados_filtrados3.iloc[i]['Partida.Prevista'])
    if r>30:
        vec.append(1)
    else:
        vec.append(0)
print(vec)
```

```python
dados_filtrados4 = dados_filtrados3
dados_filtrados4["Atrasos"] = vec
display(dados_filtrados4)
```
